در ابتدا موارد مورد نیاز را ایمپورت میکنیم، سپس توابعی برای محاسبه psnr و قطعه قطعه کردن تصویر پیاده میکنیم و در مرحله اخر با انجام فرایند ها و محاسبه زمان و psnr آن ها نتیجه گیری خود را انجام میدهیم.

```python
import matplotlib.pyplot as plt
from numpy.lib.stride_tricks import as_strided as ast
import numpy
import cv2
import time
```

حال تابعی برای محاسبه psnr پیاده میکنیم.

```python
def calculatePSNR(FirstImage , SecondImage):
    imageSize = FirstImage.shape
    MSEArray = (FirstImage - SecondImage) ** 2

    MSE = MSEArray.sum() / (imageSize[0] * imageSize[1])
    if(MSE == 0):
        return "infinity"
    PSNR = 10 * numpy.log((255 ** 2) / MSE)

    return PSNR
```

در این قسمت تابعی برای تکه تکه کردن تصویر پیاده میکنیم

```python
def block_view(A, block= (3, 3)):
    shape= (int(A.shape[0]/ block[0]), int(A.shape[1]/ block[1]))+ block
    strides= (block[0]* A.strides[0], block[1]* A.strides[1])+ A.strides
    return ast(A, shape= shape, strides= strides)
```

در این مرحله تابع block_proc را تعریف میکنیم که با استفاده از تابع بالایی تصویر مورد نظر ما را بلوک بندی میکند و سپس تابع مورد نظر ما را بر روی بلوک ها اعمال میکند.

```python
def blkproc(I,block_size,fun):
    J = numpy.copy(I)
    BJ = block_view(J,block_size)
    for n in range(BJ.shape[0]):
        for m in range(BJ.shape[1]):
            BJ[n,m] = fun(BJ[n,m])
    return J
```

حال ماسک مورد نظر سوال را ایجاد میکنیم

```python
def returnMask(size):
    return [[1 if i < (size / 2) and y < (size / 2) else 0 for i in range(size)] for y in range(size) ]

mask8 = returnMask(8)
plt.imshow(mask8, cmap="gray")
plt.title("mask")
plt.show()    
```

حال تابعی تعریف میکنیم که پس از اعمال dct با استفاده از ماسک مورد نظر این بلوک را تغیر میدهیم، سپس معکوس dct را اعمال میکنیم.

```python
def process(image):
    image = numpy.float32(image) / 255.0
    dctImage = cv2.dct(image)
    mask = returnMask(len(image))
    filteredImage = dctImage * mask
    changedImage = cv2.idct(filteredImage)
    changedImage = numpy.uint8(changedImage * 255)
    
    return changedImage
```

حال تصویر mandril را میخوانیم

```python
mandril = cv2.imread('../mandril.tiff',0)

plt.imshow(mandril, cmap="gray")
plt.title("mandril")
plt.show()    
```

حال تصویر را به تکه های 8 تایی تقسیم میکنیم و کار های مربوطه بر روی آن انجام میدهیم

```python
start_time_8_block = time.time()
processed_mandril_8_block = blkproc(mandril , (8 , 8) , process)
processed_mandril_8_block_psnr = calculatePSNR(mandril , processed_mandril_8_block)
spend_time_8_block = time.time() - start_time_8_block

plt.imshow(processed_mandril_8_block, cmap="gray")
plt.title("mandril_8_blocks")
plt.show()  
print("psnr: "+ str(processed_mandril_8_block_psnr))
print("time: " + str(spend_time_8_block))
```

حال همین کار را با بلوک بندی کردن با اندازه 16 انجام میدهیم

```python
start_time_16_block = time.time()
processed_mandril_16_block = blkproc(mandril , (16 , 16) , process)
processed_mandril_16_block_psnr = calculatePSNR(mandril , processed_mandril_16_block)
spend_time_16_block = time.time() - start_time_16_block 

plt.imshow(processed_mandril_16_block, cmap="gray")
plt.title("mandril_16_blocks")
plt.show()  
print("psnr: "+ str(processed_mandril_16_block_psnr))
print("time: " + str(spend_time_16_block))
```

حال همین کار را با اندازه 64 انجام میدهیم

```python
start_time_64_block = time.time()
processed_mandril_64_block = blkproc(mandril , (64 , 64) , process)
processed_mandril_64_block_psnr = calculatePSNR(mandril , processed_mandril_64_block)
spend_time_64_block = time.time() - start_time_64_block


plt.imshow(processed_mandril_64_block, cmap="gray")
plt.title("mandril_64_blocks")
plt.show()  
print("psnr: "+ str(processed_mandril_64_block_psnr))
print("time: " + str(spend_time_64_block))
```

همانطور که میبینیم با بلوک بندی تصویر و اعمال ماسک مورد نظر بر روی هر بلوک و باز گرداندن آن، نتایج فوق حاصل شد. با مشاهده نتایج به این نتیجه میرسیم، اندازه بلوک بندی زیاد تاثیری در کیفیت نداشته و همانطور که میبینیم کیفیت تصویر با بلوک بندی با اندازه 8 درمقابل کیفیت تصویر با بلوک بندی به اندازه 16 یا 64 تفاوت چندانی نمیکند. ولی از نظر زمان اجرا فرایند انجام شده با اندازه بلوک 64، زمان کمتری نسبت به فرایند انجام شده با اندازه بلوک 16 یا 8 دارد، البته این تفاوت نیز محسوس نیست و در حد چند دهم ثانیه است ولی امکان دارد در تصاویر با اندازه های بزرگ قابل توجه شود.

# سوال اول،بخش دوم 

تابعی برای محاسبه مقدار sharpness ایجاد میکنیم

```python
def sharpness_calculator(image):
    return cv2.Laplacian(image, cv2.CV_64F).var()
```

در این قسمت تصویر مندریل را میخوانیم و نشان میدهیم

```python
mandril2 = cv2.imread('../mandril.tiff',0)

mandril2_copy = numpy.copy(mandril2)

plt.imshow(mandril2, cmap="gray")
plt.title("mandril")
plt.show()    
```

 در این مرحله تصویر mandril را به بلوک هایی با سایز 64 تقسیم میکنیم، و باقی عملیات گفته شده در سوال را انجام میدهیم

```python
tic = time.time()

image_blocked_64_array = block_view(mandril2_copy , (64 , 64))
result_64 = list(map(lambda row: list(map(sharpness_calculator, row)), image_blocked_64_array))
reshaped_result_64 = numpy.reshape(result_64 , (1, (len(result_64) * len(result_64[0]))))[0]
sorted_64_sized_result = sorted(reshaped_result_64, reverse=True)

slice_64_index = sorted_64_sized_result[int(len(sorted_64_sized_result) / 10 + 1)]

row_index= 0
col_index = 0
for row in image_blocked_64_array:
    for col in row:
        if(cv2.Laplacian(col, cv2.CV_64F).var() <= slice_64_index):
            image_blocked_64_array[row_index][col_index] = process(image_blocked_64_array[row_index][col_index])
        else:
            image_blocked_16_array = block_view(col , (16 , 16))
            result_16 = list(map(lambda row: list(map(sharpness_calculator, row)), image_blocked_16_array))
            reshaped_result_16 = numpy.reshape(result_16 , (1, (len(result_16) * len(result_16[0]))))[0]
            sorted_16_sized_result = sorted(reshaped_result_16 , reverse=True)
            
            slice_16_index = sorted_64_sized_result[int(len(sorted_16_sized_result) / 10 + 1)]
            
            row2_index = 0
            col2_index = 0
            for row2 in image_blocked_16_array:
                for col2 in row2:
                    if(cv2.Laplacian(col2, cv2.CV_64F).var() <= slice_16_index):
                        image_blocked_16_array[row2_index][col2_index] = process(image_blocked_16_array[row2_index][col2_index])
                    else:
                        image_blocked_16_array[row2_index][col2_index] = blkproc(image_blocked_16_array[row2_index][col2_index] , (8 , 8) , process)
                        
                    col2_index = col2_index + 1
                    
                col2_index = 0
                row2_index = row2_index + 1
            
            short_result = numpy.zeros((64 , 64))

            short_block_height = 4
            short_block_width = 4

            for i in range(short_block_width):
                for j in range(short_block_height):
                    short_result[j * 16 : (j + 1) * 16 , i * 16 : (i + 1) * 16] = image_blocked_16_array[j, i]
            
            image_blocked_64_array[row_index][col_index] = short_result
        col_index = col_index + 1
    
    col_index = 0
    row_index = row_index +1
            
result = numpy.copy(mandril2)

block_height = 8
block_width = 8

for i in range(block_width):
        for j in range(block_height):
            result[j * 64 : (j + 1) * 64,i * 64 : (i + 1) * 64] = image_blocked_64_array[j, i]
    

spend_time = time.time() - tic
```

حال تصویر نتیجه شده، مقدار psnr و زمان مصرف شده را نشان میدهیم

```python
processed_mandril2_psnr = calculatePSNR(mandril2 , mandril2_copy)

plt.imshow(mandril2_copy, cmap="gray")
plt.title("mandril2")
plt.show()  
print("psnr: "+ str(processed_mandril2_psnr))
print("time: " + str(spend_time))
```
